```python
import pandas as pd
import bar_chart_race as bcr
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
df.head()
```

```python
df = df.drop(columns=["Lat","Long"])
```

```python
# nouvelle colonne "Location" qui combine "Province/State" et "Country/Region"
df['Location'] = df[['Province/State','Country/Region']].apply(lambda x: ', '.join(x.dropna()),axis=1)
df.sample(3)
```

```python
df = df.drop(columns=["Province/State", "Country/Region"])
```

```python
# mettre la colonne 'Location' en premiere position
cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
```

```python
# transposer ligne/colonne pour que le pays soit le nom de la conne et que chaque date soit une ligne
df_transposed = df.T
df_transposed
```

```python
# la fonction transpose a mis les pays dans la premiere ligne
# on veut que ce soit le nom des colonnes
df_transposed.columns = df_transposed.iloc[0].to_list()
df_transposed.columns
```

```python
df_transposed
```

```python
# il faut supprimer la premiere ligne que est le nom des pays
df_transposed = df_transposed.iloc[1:]
df_transposed
```

```python
# labiliser l'index avec le nom "Date"
df_transposed.index.names = ["Date"]
df_transposed
```

```python
# on ne s'interresse qu'aux pays suivant
cols = ["France","US","Germany","Italy","United Kingdom","Brazil","Russia","Spain"]
df_subset = df_transposed[cols]
df_subset
```

```python
# creer un datafrane où chaque ligne s'aditionne à la précédente
# mettre toutes les types est numeric 
df_cum_sum = df_subset.cumsum()
df_cum_sum = df_cum_sum.apply(pd.to_numeric)
```

```python
# mettre l'index (Date) en type datetime
df_cum_sum.index = pd.to_datetime(df_cum_sum.index)
```

```python
# plotting
bcr.bar_chart_race(
    df_cum_sum,
    title="Evolution du Covid19",
    interpolate_period=True,  
    period_fmt="%b %-d, %Y"
)
# filename="covid.mp4"
```
